```python
# we need to extact the  libraries
# using the import function

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# creating a data frame to read the data

df = pd.read_csv('Telco_Cusomer_Churn.csv')
df
```

```python
# we are replacing blanks with 0 as tenure is 0 and no total charges are recorded
df["TotalCharges"]=df["TotalCharges"].replace(" ","0")
df["TotalCharges"]= df["TotalCharges"].astype("float")
```

```python
# converted 0 and 1 values of SeniorCitizen to yes/no to make it easier to understand
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"

df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
```

```python
ax =sns.countplot(x= "Churn" , data = df)
ax.bar_label(ax.containers[0])
plt.title("Count of customers by Churn")
plt.show()
```

```python
gb = df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gb['Churn'] , labels = gb.index , autopct = "%1.2f%%")
plt.title("Churn Percentage")
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.countplot(x = "gender" , data = df, hue = "Churn")
plt.title("Churn Distribution by Gender")
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.countplot(x="SeniorCitizen", data=df, hue="Churn")

plt.title("Churn Distribution by Senior Citizen Status")
plt.xticks([0, 1], ['Non-Senior', 'Senior']) # Renames 0 and 1 for clarity
plt.show()
```

```python
plt.figure(figsize=(4,4))
ax = sns.countplot(x="SeniorCitizen", data=df, hue="Churn")
ax.bar_label(ax.containers[0])
plt.title("Churn Distribution by Senior Citizen Status")
plt.xticks([0, 1], ['Non-Senior', 'Senior']) # Renames 0 and 1 for clarity
plt.show()
```

```python
# 1. Calculate the percentage distribution
# We group by SeniorCitizen and Churn, get the counts, and normalize them by row
gb_pct = df.groupby('SeniorCitizen')['Churn'].value_counts(normalize=True).unstack() * 100

# 2. Plotting the stacked bar chart
ax = gb_pct.plot(kind='bar', stacked=True, figsize=(6, 6), color=['#4C72B0', '#C44E52'])

# 3. Customizing the labels and title
plt.title("Churn Distribution by Senior Citizen Status (Percentage)")
plt.xlabel("Senior Citizen Status")
plt.ylabel("Percentage (%)")
plt.xticks([0, 1], ['Non-Senior', 'Senior'], rotation=0)
plt.legend(title='Churn', bbox_to_anchor=(1.05, 1), loc='upper left')

# 4. Adding the % labels inside the bars
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy() 
    if height > 0:
        ax.text(x + width/2, 
                y + height/2, 
                f'{height:.1f}%', 
                horizontalalignment='center', 
                verticalalignment='center',
                color='white',
                fontsize=11,
                weight='bold')

plt.tight_layout()
plt.show()
```

```python
# List of columns you want to plot
cols = ['PhoneService', 'MultipleLines', 'InternetService', 
        'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 
        'TechSupport', 'StreamingTV', 'StreamingMovies']

# Set up the figure and axes (3 rows, 3 columns)
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 15))

# Flatten the axes array for easy iteration
axes = axes.flatten()

for i, column in enumerate(cols):
    # Create the countplot on the specific axis
    sns.countplot(x=column, data=df, ax=axes[i], hue="Churn")
    
    # Add titles and labels
    axes[i].set_title(f'Churn by {column}')
    axes[i].set_xlabel("") # Removes redundant x-labels for a cleaner look
    
    # Optional: Add bar labels
    for container in axes[i].containers:
        axes[i].bar_label(container)

# Adjust layout so titles don't overlap
plt.tight_layout()
plt.show()
```
